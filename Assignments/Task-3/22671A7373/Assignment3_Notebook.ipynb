```python
# 2. Import libraries
import cv2
import matplotlib.pyplot as plt

# helper function to show image
def show_img(img, cmap=None, title="Image"):
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
# 3. Read Image
img = cv2.imread("sample.jpg")   # replace with your image path
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
show_img(img_rgb, title="Original Image")
```

```python
# 4. Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_img(gray, cmap="gray", title="Grayscale Image")
```

## Grayscale Conversion
The image is converted into grayscale to simplify processing by reducing it to intensity values.


```python
# 5. Gaussian Blur
blur = cv2.GaussianBlur(img_rgb, (7,7), 0)
show_img(blur, title="Gaussian Blur")
```

## Gaussian Blur
We apply Gaussian Blur to smooth the image and reduce noise.


```python
# 6. Edge Detection
edges = cv2.Canny(img, 100, 200)
show_img(edges, cmap="gray", title="Edges (Canny)")
```

## Edge Detection
Using the Canny algorithm, we detect edges in the grayscale image.


```python
# 7. Rotate
(h, w) = img.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 90, 1.0)
rotated = cv2.warpAffine(img_rgb, M, (w, h))
show_img(rotated, title="Rotated 90°")
```

## rotation 
The imageis rotated 90 degrees

```python
# 8. Resize
resized = cv2.resize(img_rgb, (200, 200))
show_img(resized, title="Resized (200x200)")
```

## Resizing
The image is resized to 200x200 pixels.
